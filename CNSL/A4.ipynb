```python
# If the parity of the data that is to be send is Odd we return 1 else 0.
# We reserve the 1st bit of 'encoded_data_bits' for parity check.

encoded_data_bits = [0,0,0,0,0,0,0]

def final_data(corrected_data_bits, correct_status):
    if correct_status == 1:
        print("After self-correction, the data is: {}".format(corrected_data_bits))


def parity_check(test_data):
    encoded_data_bits[2] = test_data[4] & test_data[5] & test_data[6]
    encoded_data_bits[1] = test_data[3] & test_data[5] & test_data[6]
    encoded_data_bits[0] = test_data[3] & test_data[4] & test_data[6]

def hamming_code(sender_data):
    error_code = 0
    parity_check(sender_data)
        
    print("Data sent is {}\n".format(sender_data))
    
    received_data = [0,0,0,0,0,0,0]
    for i2 in range(len(received_data)):
        check_data = int(input("Enter received bit: "))
        if check_data == 1 or check_data == 0:
            received_data[i2] = check_data
        else:
            print("\nOnly binary digits are accepted.\nError code exit -2.")
            error_code = -2
            break
            
    if error_code == 0:
        print("\nReceived data is: {}".format(received_data))
        receiver_check(sender_data, received_data)
    
    
    
def receiver_check(sent_data, received_data):
    error_pos = []
    error_bit = 0
    for i in range(len(sent_data)):
        if sent_data[i] != received_data[i]:
            print("\nError found at position {}.".format(i))
            error_pos.append(i)
            error_bit += 1
        
    if error_bit != 0:
        print("\n{} error/s were found.".format(error_bit))
        self_correction(error_pos, received_data)
    else:
        print("\nReceived data is error free.")
        final_data(received_data, 0)

        
        
def self_correction(positions_list, received_data):
    for i3 in positions_list:
        if received_data[i3] == 1:
            received_data[i3] = 0
        
        else:
            received_data[i3] = 1
            
    final_data(received_data, 1)
    

def sending_data(raw_data):
    error_code = 0
    print("We will be sending 4-bit data.")
    for i in range(1,5):
        get_data = int(input("Enter data bit: "))
        if get_data == 1 or get_data == 0:
            raw_data[-i] = get_data
        else:
            print("\nOnly Single binary digits are accepted.\nError code exit -1.")
            error_code = -1
            break
    if error_code == 0:
        print("\nUnprocessed data: {}".format(raw_data))
        hamming_code(raw_data)
        
sending_data(encoded_data_bits)
```
